```python
# from sklearn.svm import SVC
# from sklearn.model_selection import train_test_split
# from sklearn.metrics import accuracy_score

# # Example dataset of emoji strings and labels
# emoji_strings = X
# labels = data['label']  # Binary classification labels

# # Step 1: One-Hot Encoding of the emoji strings
# # Split emoji strings into individual emojis (characters)
# emoji_lists = [list(emoji_string) for emoji_string in emoji_strings]

# # Initialize OneHotEncoder
# onehot_encoder = OneHotEncoder()

# # Fit and transform the data into one-hot encoded vectors
# X_onehot = onehot_encoder.fit_transform(emoji_lists)

# # Step 2: Split data into training and testing sets
# X_train, X_test, y_train, y_test = train_test_split(X_onehot, labels, test_size=0.2, random_state=42)

# # Step 3: Initialize and train the SVM classifier
# svm_clf = SVC(kernel='linear')
# svm_clf.fit(X_train, y_train)

# # Step 4: Make predictions on the test set
# y_pred = svm_clf.predict(X_test)

# # Step 5: Calculate and print the accuracy
# accuracy = accuracy_score(y_test, y_pred)
# print(f"Accuracy: {accuracy * 100:.2f}%")



# valdata=pd.read_csv('C:/Users/gupil/Downloads/mini-project-1/mini-project-1/datasets/valid/valid_emoticon.csv')

# emoji2 = list(valdata['input_emoticon'])
# labels2 = valdata['label']  # Binary classification labels

# # Step 1: One-Hot Encoding of the emoji strings
# # Split emoji strings into individual emojis (characters)
# emoji_lists2 = [list(emoji_string) for emoji_string in emoji2]

# # Initialize OneHotEncoder
# onehot_encoder2 = OneHotEncoder()

# # Fit and transform the data into one-hot encoded vectors
# X_onehot2 = onehot_encoder2.fit_transform(emoji_lists2)

# print(X_onehot2.shape)

# # Predict on the validation data
# y_val_pred = svm_clf.predict(X_onehot2)

# # Evaluate the model
# accuracy = accuracy_score(labels2, y_val_pred)
# print(f"Validation Accuracy: {accuracy:.2f}")

moji_strings = X  # Your training emojis
labels = data['label']  # Binary classification labels

# Step 1: One-Hot Encoding of the emoji strings (Training data)
# Split emoji strings into individual emojis (characters)
emoji_lists = [list(emoji_string) for emoji_string in emoji_strings]

# Initialize OneHotEncoder
onehot_encoder = OneHotEncoder()  # Set sparse=False to get an array output

# Fit the encoder on the training data
X_onehot = onehot_encoder.fit_transform(emoji_lists)

# Step 2: Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_onehot, labels, test_size=0.2, random_state=42)

# Step 3: Initialize and train the SVM classifier
svm_clf = SVC(kernel='linear')
svm_clf.fit(X_train, y_train)

# Step 4: Make predictions on the test set
y_pred = svm_clf.predict(X_test)

# Step 5: Calculate and print the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Training Set Accuracy: {accuracy * 100:.2f}%")

# Load your validation dataset
valdata = pd.read_csv('C:/Users/gupil/Downloads/mini-project-1/mini-project-1/datasets/valid/valid_emoticon.csv')

emoji2 = list(valdata['input_emoticon'])
labels2 = valdata['label']  # Binary classification labels

# Step 1: One-Hot Encoding of the emoji strings (Validation data)
# Split emoji strings into individual emojis (characters)
emoji_lists2 = [list(emoji_string) for emoji_string in emoji2]

# Transform the validation data using the fitted OneHotEncoder
X_onehot2 = onehot_encoder.transform(emoji_lists2)  # Use the same encoder

# Predict on the validation data
y_val_pred = svm_clf.predict(X_onehot2)

# Evaluate the model
accuracy_val = accuracy_score(labels2, y_val_pred)
print(f"Validation Accuracy: {accuracy_val*100:.2f}%")

```

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize and train the Decision Tree classifier
dt_clf = DecisionTreeClassifier(random_state=42)
dt_clf.fit(X_train, y_train)

# Predict and calculate accuracy
y_pred = dt_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Decision Tree Accuracy: {accuracy * 100:.2f}%")
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize and train the Random Forest classifier
rf_clf = RandomForestClassifier(n_estimators=150, min_samples_leaf=2,min_samples_split=5,max_features='log2',random_state=42)
rf_clf.fit(X_train, y_train)

# Predict and calculate accuracy
y_pred = rf_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Random Forest Accuracy: {accuracy * 100:.2f}%")
```
